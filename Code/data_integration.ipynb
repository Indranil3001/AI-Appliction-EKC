## GDP Data Prepration

```python
col_to_keep_in_gdp= ['Country Name','Country Code', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997',
               '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', 
              '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', 
              '2018', '2019', '2020','2021','2022','2023','2024']
df_gdp_absolute = df_gdp_absolute[col_to_keep_in_gdp]
df_gdp_absolute.rename(columns={'Country Name': 'Country', 'Country Code': 'ISO3'}, inplace=True)
df_gdp_absolute.head()
```

## Changing GDP and GHG to long format

## GDP Data in long format from 1990 onwards

```python
#GDP
year_cols_gdp = [col for col in df_gdp_absolute.columns if col.isdigit() and int(col) >= 1990]
df_gdp_long = df_gdp_absolute.melt(
    id_vars=['Country', 'ISO3'],
    value_vars=year_cols_gdp,
    var_name='Year',
    value_name='GDP_USD'
)
df_gdp_long.head()
```

```python
df_gdp_long= df_gdp_long.sort_values(by=['Country','Year']).reset_index(drop=True)
df_gdp_long.head(50)
```

## GHG Data in long format from 1990 onwards

```python
#GHG
year_cols_ghg = [col for col in df_ghg_filtered.columns if col.isdigit() and int(col) >= 1990]
df_ghg_long = df_ghg_filtered.melt(
    id_vars=['Country', 'ISO3'],
    value_vars=year_cols_ghg,
    var_name='Year',
    value_name='GHG_Emissions'
)
df_ghg_long.head(100)
```

```python
df_ghg_long= df_ghg_long.sort_values(by=['Country','Year']).reset_index(drop=True)
df_ghg_long.head(100)
```
